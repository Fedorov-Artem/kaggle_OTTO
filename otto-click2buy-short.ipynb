# Click2buy short co-visitation matrix
This notebook prepares click-to-buy short co-visitation matrix. It is used to calculate a feature for both carts and orders models.
Difference with click2buy matrix is that here  only events with shorter time difference between them are counted and the weight value declines much faster as time difference grows.
## Imports and definitions

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import gc
from humanize import naturalsize

# functions and classes common for several notebooks of current project
import otto_common
```

```python
class CalculateCovisitationMatrixClick2BuyShort(otto_common.CalculateCovisitationMatrix):
    '''
    Child class with logic specific to this type of co-visitation matrix.
    It counts all aids that were either carted or ordered within a certain time after some aid was clicked.
    The weight value is calculated in a way that makes pairs of events with smaller time difference more important.
    Difference with click2buy matrix is that here time difference is limited to much shorter time and the weight value
    declines much faster in time.
    '''

    def calculate_weights(self, df_i):
        df_i = df_i.sort_values(['session','ts'],ascending=[True,False])
        df_i = df_i.reset_index(drop=True)
        df_i = df_i.merge(df_i,on='session')
        df_i = df_i.loc[ (df_i.ts_y >= df_i.ts_x) & (df_i.type_y > 0)]
        df_i['time_diff'] = df_i.ts_y - df_i.ts_x
        df_i = df_i.loc[ (df_i.time_diff).abs() < 2 * 60 * 60]
        df_i = df_i.groupby(['session','aid_x', 'aid_y']).agg({'time_diff':'min'}).reset_index()
        df_i['wgt'] = (5 * 60)/(5 * 60 + df_i.time_diff)
        df_i = df_i[['aid_x','aid_y','wgt']]
        df_i = df_i.loc[df_i['aid_x'].isin(self.aid_list)]
        df_i.wgt = df_i.wgt.astype('float32')
        df_i = self.groupby_reset_and_reduce(df_i)
        return df_i
```

## Co-visitation matrix for cross-validation dataset

```python
# Make a list of AIDs, that appear in one of cross-validation datasets.
# aid_x is filtered to only include aids from the list, while aid_y is not filtered
# This reduces the size of the resulting matrix.
trunked_sessions_path = '/kaggle/input/otto-prepare-cv/cv_inputs.parquet'
trunked_sessions_path2 = '/kaggle/input/otto-prepare-cv/cv_inputs2.parquet'
cv_list = otto_common.build_aid_list(trunked_sessions_path, trunked_sessions_path2)
```

```python
click_data_cv = '/kaggle/input/otto-prepare-cv/cv_train.parquet'
n_splits = 200
n_splits_concat = 25
covisitation_click2buy_cv = CalculateCovisitationMatrixClick2BuyShort(n_splits, n_splits_concat, cv_list)
df_matrix_cv = covisitation_click2buy_cv.generate_covisitation_matrix(click_data_cv)
```

```python
# Check dataframe size.
size = df_matrix_cv.memory_usage(deep='True').sum()
print(naturalsize(size))
df_matrix_cv.to_parquet('matrix_click2buy_cv_short.parquet')

del df_matrix_cv, cv_list, covisitation_click2buy_cv
gc.collect()
```
